# Detector check

```python
import cv2 
import matplotlib.pyplot as plt 
img=cv2.imread("tests/page.png")
plt.imshow(img)
plt.show()
```

```python
# initialization
from BengaliaiOCR import Detector
detector=Detector(use_gpu=False)
# getting word boxes
word_boxes=detector.get_word_boxes(img)
# getting line boxes
line_boxes=detector.get_line_boxes(img)
# getting crop with either of the results
crops=detector.get_crops(img,word_boxes)

for crop in crops[:3]:
    plt.imshow(crop)
    plt.show()
```
